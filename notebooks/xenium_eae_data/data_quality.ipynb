```python
import matplotlib.pyplot as plt

# Genes per cell
sc.pl.violin(adata, ["n_genes_by_counts"], jitter=0.4, multi_panel=True, size = 0.1)

# Counts per cell
sc.pl.violin(adata, ["total_counts"], jitter=0.4, multi_panel=True, size = 0.1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplots of n_genes per sample
plt.figure(figsize=(25,4))
sns.boxplot(data=adata.obs, x="sample_id", y="total_counts")
plt.xticks(rotation=90)
plt.title("Total counts per cell by sample")
plt.show()

# Boxplots of total_counts per condition
plt.figure(figsize=(6,4))
sns.boxplot(data=adata.obs, x="condition", y="total_counts")
plt.title("Total counts per cell by condition")
plt.show()

# Boxplots of n_genes by course
plt.figure(figsize=(6,4))
sns.boxplot(data=adata.obs, x="course", y="n_genes_by_counts")
plt.title("Genes per cell by disease course")
plt.xticks(rotation=90)
plt.show()

# Boxplots of n_genes by course
plt.figure(figsize=(6,4))
sns.boxplot(data=adata.obs, x="course", y="total_counts")
plt.title("Total counts per cell by disease course")
plt.xticks(rotation=90)
plt.show()
```

```python
# Genes vs counts per cell
sns.scatterplot(data=adata.obs, x="total_counts", y="n_genes_by_counts", hue="course", s=5, alpha=0.5)
plt.title("Genes vs counts per cell")
plt.show()
```
